```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import pickle

# Load cleaned data (ensures indices match)
cleaned_df = pd.read_csv("cleaned_data.csv")
print(f"Loaded {len(cleaned_df)} rows from cleaned_data.csv")
print("Original columns:", cleaned_df.columns.tolist())

```

```python
# Define categorical columns for encoding
cat_cols = ["main_city", "cuisine"]

# Initialize and fit OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_cats = encoder.fit_transform(cleaned_df[cat_cols])

# Get feature names for encoded columns
feature_names = encoder.get_feature_names_out(cat_cols)

# Identify numeric columns (exclude categorical)
num_cols = ["rating", "rating_count", "cost"]  # Adjust based on your dataset
num_cols = [col for col in num_cols if col in cleaned_df.columns]
```

```python
# Create encoded DataFrame with matching indices
encoded_cat_df = pd.DataFrame(encoded_cats, columns=feature_names, index=cleaned_df.index)
encoded_df = pd.concat([cleaned_df[num_cols], encoded_cat_df], axis=1)
```

```python
# Verify all features are numerical
print("Data types after encoding:")
print(encoded_df.dtypes)
print("All numerical:", encoded_df.select_dtypes(include=['number']).shape[1] == encoded_df.shape[1])
```

```python
# Save encoded dataset
encoded_df.to_csv("encoded_data.csv")
print("✅ Saved encoded_data.csv (ML ready)")

# Save encoder for future use (Streamlit app)
with open("encoder.pkl", "wb") as f:
    pickle.dump(encoder, f)
print("✅ Saved encoder.pkl")

# Verify indices match
cleaned_indices = pd.read_csv("cleaned_data.csv", index_col=0).index
encoded_indices = pd.read_csv("encoded_data.csv", index_col=0).index
print("✅ Indices match:", (cleaned_indices.equals(encoded_indices)))
```
