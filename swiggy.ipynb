```python
swiggy=['Bangalore','Delhi','Pune','chennai','Hyderabad','Kolkata','Mumbai','Ahmedabad','Chandigarh','Lucknow']
filtered_swiggy=swiggy_[swiggy_['city_final'].isin(swiggy)]
filtered_swiggy
```

```python
filtered_swiggy.to_csv("cleaned_data.csv",index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

cleaned_df = filtered_swiggy
cat_cols = ["city_final", "cuisine"]
num_cols = ["rating", "rating_count", "cost"]

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_cats = encoder.fit_transform(cleaned_df[cat_cols])
feature_names = encoder.get_feature_names_out(cat_cols)

# Reset indices to match exactly
encoded_cat_df = pd.DataFrame(encoded_cats, columns=feature_names,
                             index=cleaned_df.index).reindex(cleaned_df.index)
encoded_df = pd.concat([cleaned_df[num_cols], encoded_cat_df], axis=1)

encoded_df.to_csv("encoded_data.csv", index=False)
print("✅ encoded_df shape:", encoded_df.shape)  
```

```python
import pickle

#Save encoder for future use (Streamlit app)
with open("encoder.pkl", "wb") as f:
  pickle.dump(encoder, f)
  print("✅ Saved encoder.pkl")
```

```python
from sklearn.cluster import KMeans
import pandas as pd

# Load data
cleaned_df = pd.read_csv('cleaned_data.csv')
encoded_df = pd.read_csv('encoded_data.csv')

# K-Means Clustering 
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
clusters = kmeans.fit_predict(encoded_df)

# ADD MAPPING INDICES 
cleaned_df['cluster'] = clusters  # Maps clusters back by index
cleaned_df.to_csv('final_data.csv', index=False)

print("✅ Clusters mapped! Use: cleaned_df[cleaned_df['cluster'] == 0]")
```
